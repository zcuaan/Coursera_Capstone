```python
from bs4 import BeautifulSoup
import requests
import numpy as np
import pandas as pd
import pickle
```

```python
pickle_in = open("df.pickle","rb")
check = pickle.load(pickle_in)
```

```python
address = check['Address']
```

```python
def hasNumbers(inputString):
    return any(char.isdigit() for char in inputString)
```

```python
# Post codes
postcodes = []
posttown = []
main = []
for i in range(len(address)):
    txt = address[i].split(',')
    print(i, txt)
    if hasNumbers(txt[-1]):
        postcodes.append(txt[-1].strip())
        if len(txt) == 1:
            posttown.append('NA')
            main.append(txt[0])
        else:
            posttown.append(txt[-2].strip())
            main.append(','.join(txt[:-2]))
    else:
        postcodes.append('NA')
        posttown.append(txt[-1])
        main.append(','.join(txt[:-1]))
```

```python
df = pd.DataFrame({'Main': main, 'Post Town': posttown, 'Post Code': postcodes})
```
